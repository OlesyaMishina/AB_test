```python

n1=n2=15550
m1=164
m2=228
conv1=round(m1/n1,4)
conv2=round(m2/n2, 4)
conv1, conv2
```

```python
d=(((conv1*(1-conv1))/n1+(conv2*(1-conv2))/n2)**0.5)
d
print(f'Доверительный интервал для разницы конверсий: {round((conv2-conv1)-1.96*d, 4)}; {round((conv2-conv1)+1.96*d, 4)}')
```

Доверительный интервал не проходит через 0 и расположен в положительной части оси, это говорит о том, что статистическая разница есть, второй вариант эксперимента более успешен, что подтверждается и рассчетом через калькулятор.
